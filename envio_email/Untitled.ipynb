```python
import pandas as pd
tabela = pd.read_excel('Vendas.xlsx')
```

```python
pd.set_option('display.max_columns', None)

faturamento = tabela[['ID Loja','Valor Final']]
faturamento = faturamento.groupby('ID Loja').sum()

quantidade = tabela[['ID Loja','Quantidade']]
quantidade = quantidade.groupby('ID Loja').sum()

ticket_medio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns={0:'Ticket Médio'})

display(ticket_medio)
```

```python
import win32com.client as win32
outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)
mail.To = 'luana.oliveira-ext@via.com.br'
mail.Subject = 'Teste Python'
mail.HTMLBody = f'''<p>Prezados,&nbsp;</p>
<p>Segue o relat&oacute;rio de Vendas por cada Loja.</p>
{faturamento.to_html(formatters={'Valor Final': 'R$ {:,.2f}'.format}).replace('.','*').replace(',','.').replace('*',',')}
<p>Faturamento: <br> {faturamento.to_html()} <br>
Quantidade Vendida: <br> {quantidade.to_html()} <br>
{ticket_medio.to_html(formatters={'Ticket Médio': 'R$ {:,.2f}'.format})}
Ticket M&eacute;dio dos produtos por cada loja: <br> {ticket_medio.to_html()} <br>
Qualquer d&uacute;vida estou a disposi&ccedil;&atilde;o.&nbsp;</p>
<p>Att.</p>
<p><strong>Andreia Ferreira</strong><br><strong><span style="color: rgb(184, 49, 47);">Analista de Desenvolvimento</span></strong></p>
<p><br></p>
<p><br></p>'''
mail.Send()
```
